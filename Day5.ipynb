```python
#While loop with conditions
print("find how many numbers from 1 to n are even & odd numbers")
n=int(input("enter a number n: "))
i=1
even=0
odd=0
while i<=n:
    if i%2==0:
        print("Even ",i)
        even+=1
    else:
        print("Odd",i)
        odd+=1
    i+=1
print("total even-> ",even)
print("total odd-> ",odd)
```
