```python
#3calculator
x0=int(input("enter 1st number"))
y0=int(input("enter 2nd number "))
opr=input("enter operator")
def clc_optn(x,y,op):
    if op=="+":
        return x+y
    elif op=="-":
        return x-y
    elif op=="*":
        return x*y
    elif op=="/":
        return x/y
    else:
        return "invalid op"

print(clc_optn(x0,y0,opr))
```
