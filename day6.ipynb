```python
#function with keyword parameters
def finder(x,y):
    print("value of x:",x)
    print("value of y:",y)
```

```python
finder(4,5)
```

```python
finder(x=5,4)
```

```python
finder(x=5,y=4)
```

```python
def check(x,y):
    if x>y:
        return x
    return y
```

```python
res=check(4,8)
```

```python
res
```

```python
def divide(a,b):
    return a//b

print(divide(4,2))
```

```python
#default parameters
def pow(x=2,y=3):
    print(x**y)
```

* "args" =>tuple:many values

```python
def show(*x):
    print(x)

show(4,2,5,6,"hello")
```

**"kwargs" =>dictionary: labeled val

```python
def dict_show(**x):
    print(x)
dict_show(name="grik",rollNo=4,skill="python")
```

lambda function: quick one-line function without name

```python
square=lambda x:x**x
square(4)
```
