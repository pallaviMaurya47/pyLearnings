```python
def double_no(x):
    return x*2

numbers=[3,4,2,5,6]
res=list(map(double_no,numbers))
print(res)
```

```python
res=list(map(lambda x:x*x , [1,2,3,4,5]))
print(res)
```

```python
#Filter()
def is_even(x):
    return x%2==0

result=list( filter(is_even, [23,40,22,84,53]))
print(result)
```

Sorted() function

```python
nums=[9,3,2,3,4,0,7]
print(sorted(nums))
```

```python
def word(word):
    return len(word)

words=["Tree","fruits", "match","pine,ginger"]
length=list(sorted(words, key=word))
print(length)
```

```python
#sum
def sum_numbers(*x):
    return sum(x)

# number_s=[3,2,5,7,8,3,4]
print(sum_numbers(44,2,3,5))
```
